# <font color='maroon'>The Normal distribution</font>

Continuous variables like weights, heights, ages, atmospheric pressure, metabolic rates can be modelled by a normal distribution. Additionally, given large enough samples, we find that estimates of random variables that follow other distributions eventually start to follow a normal distribution. In this lesson we learn about about the Normal Distribution, its scaling down to a standard normal distribution, and the cumulative frequency distribution. We also generate some data to visualise the shape of these distributions. 

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from math import sqrt
```

```python
%matplotlib inline
```

The Normal distribution is also called the Gaussian distribution. The shape of the distribution is symmetric and centred around the mean. Let's see how this looks with data that we will simulate using numpy: np.random.normal. 

```python
n = np.random.normal(100*0.5,           # start
                     sqrt(100*0.5*0.5), # stop
                     size=10000)        # number of points in interval
plt.hist(n,         # number of data points
         alpha=0.5, # transparency of bars
         bins=20)   # number of intervals
```

We can determine the mean and variance of the simulated data.

```python
m = n.mean()
v = n.var()
```

We plot a density curve of the simulated data.

```python
x = np.arange(-50,150,.05)
y = stats.norm.pdf(x, loc=m, # mean
                   scale=v)  # variance
#the line above generates a normal function centered about the mean, m with variance, v.
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y,'g-')
```

The normal distribution has a probability density function:

\begin{align}
f(x) &= \dfrac{1}{\sqrt(2\pi \sigma^2)}e^{-\frac{1}{2}(\frac{x - \mu}{\sigma})^2} & -\infty < x < \infty
\end{align}
where $\mu$ is the mean, standard deviation $\sigma$. A random variable $X$ is normally distributed if it can be described by this distribution.

The normal distribution is often normalized to a standard normal distribution. This has the effect of reducing the error under the curve to one.

### The standard normal distribution

The standard normal distribution has probability density function $$f(x) = \frac{1}{\sqrt(2 \pi)} e^{-\frac{1}{2}x^2},$$ with mean equal zero and variance equal 1. It is 'standard' since the mean has been subtracted from each value and then normalized by the mean. The resulting value of this standard normal distribution is referred to as a z- score. $$ z = \frac {\mu -x}{\sigma}$$. The density plot below illustrates the standard normal distribution.


```python
x = np.arange(-5,5,.05)
y = stats.norm.pdf(x, loc=0.0, # mean
                   scale=1.0)  # variance
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y,'g-')
```

## Expected value

The expected value or the mean of the random variable is the <b>mean of the normal distribution</b>.

### Cumulative Frequency Distribution

In order to answers questions like, what is the probability that an event will take on a probability greater than or equal to a certain probability, we use the cumulative density function. It is stated as follows:
$$F(x) = P(X \leq x).$$ Again, it simply asks what is the probability that the random variable $X$ will take on a value equal $x$ or less? Let's plot this curve.

```python
x = np.arange(0,10,.05)
y = stats.norm.cdf(x, loc=0.0, scale=1.0)
plt.xlabel('x')
plt.ylabel('Cumulative density, F(x)')
plt.plot(x,y,'g-')
```
